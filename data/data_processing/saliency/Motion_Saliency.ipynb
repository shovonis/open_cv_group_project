```python
baseAddress = "content/drive/Shared with me/DeepVDS/data/testVid.mp4"

#initialize saliency
saliency = None

frameCnt = 1

#Grab video
baseAddress = "testVid.mp4"
cap = cv2.VideoCapture(baseAddress)

# check that it can open
if (cap.isOpened() == False):
    print("Error opening video file")
    exit(-1)

#read in video
while(cap.isOpened()):
    ret, frame = cap.read()
    if ret == True:
        #resize frame
        frame = imutils.resize(frame, width=500)

        #if saliency hasn't been initialized do so
        if saliency is None:
            saliency = cv2.saliency.MotionSaliencyBinWangApr2014_create()
            saliency.setImagesize(frame.shape[1], frame.shape[0])
            saliency.init()

        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        (success, saliencyMap) = saliency.computeSaliency(gray)
        saliencyMap = (saliencyMap * 255).astype("uint8")

        #this is useful for when doing outside of a notebook
        # cv2.imshow("frame", frame)
        # cv2.imshow("map", saliencyMap)

        #saves each frame map in the working directory it'll be messy
        fileNm = "FrameMap" + str(frameCnt)
        cv2.imwrite(fileNm, saliencyMap)
        frameCnt += 1


        #Press Q on keyboard to quit if something is wrong
        if cv2.waitKey(0) & 0xFF == ord('q'):
            break

    else:
        break

#release the video
cap.release()

#close all frames if using the imshow
# cv2.destroyAllWindows()
```
